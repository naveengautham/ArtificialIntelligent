```python
from pathlib import Path
```

```python
from autogen import ConversableAgent
from autogen.coding import LocalCommandLineCodeExecutor
```

```python
working_dir= Path("Save_coding_doc")
working_dir.mkdir(exist_ok=True)
```

```python
executor= LocalCommandLineCodeExecutor(work_dir=working_dir)
```

```python
from Util import get_openai_api_key
OPEN_API_KEY=get_openai_api_key()
llm_config={
    "config_list":[
        {
            "model":"gpt-3.5-turbo",
            "api_key":OPEN_API_KEY
        }
    ]
}
```

```python
#agent 1
code_executor_agent= ConversableAgent(
    name="code_executor_agent",
    llm_config=False,
    code_execution_config={
        "executor":executor
    },
    human_input_mode="NEVER"
    
)
```

```python
code_writer_system_message="""You are a helpful and detail-oriented Python programming assistant. Your job is to analyze Python code line by line.

Goal:
- The code should determine whether a number is even or odd.
- Accept integer input from the user.
- Output whether the number is "Even" or "Odd".

Your task:
1. Review the code line by line and provide suggestions to improve readability, efficiency, and code quality.
2. Verify whether the logic is correct for both even and odd inputs.
3. If the code has any issues, explain why and suggest a corrected version.
4. Ensure that the code handles user input safely and gives expected output.

Respond in this format:
- **Original Line**: `<line>`
- **Feedback**: `<suggestion or observation>`
- **Improved Line** *(if any)*: `<corrected or improved line>`

At the end, provide:
- A final improved version of the code.
- Sample input and output examples.
"""
```

```python
#agent 2
code_writer_agent=ConversableAgent(
    name="code_executor_agent",
    system_message=code_writer_system_message,
    llm_config=llm_config,
    code_execution_config=False,
    max_consecutive_auto_reply=2,
    human_input_mode="NEVER"
)
```

```python
import pprint
chat_results= code_executor_agent.initiate_chat(
    code_writer_agent,message="Write a python program to find an odd an even number and display the result in an ascending format list"
)
pprint.pprint(chat_results)
```
