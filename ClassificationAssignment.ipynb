```python
import pandas as pd
```

```python
dataset= pd.read_csv("CKD.csv")
dataset= pd.get_dummies(dataset,drop_first=True)
dataset= dataset.replace({True:1,False:0})
dataset
```

```python
dataset.columns
```

```python
independent= dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
independent
```

```python
dependent= dataset[['classification_yes']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['gini','entropy'],'max_features':['auto','sqrt','log2'],'n_estimators':[50,100]}
grid= GridSearchCV(RandomForestClassifier(),param_grid,refit=True,n_jobs=-1,verbose=3,scoring='f1')
grid.fit(X_train,y_train)
```

```python
re= grid.cv_results_
```

```python
grid_pred= grid.predict(X_test)
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,grid_pred)
cm
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,grid_pred)
print(clf_report)
```

```python
#to check the f1 score value manually
from sklearn.metrics import f1_score
f1_macro= f1_score(y_test,grid_pred,average='macro')
print(f1_macro)
```

```python
#to check good model or not
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
```
