```python
def sentimentalAnalysis(data,columnName):
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    sid= SentimentIntensityAnalyzer()
    #deleting empty rows
    data.dropna(inplace=True)
    data['Scores']=data[columnName].apply(lambda columnName:sid.polarity_scores(columnName))
    data['Compound']=data['Scores'].apply(lambda comp: comp['compound'])
    data['Neutral']=data['Scores'].apply(lambda neu: neu['neu'])
    data['Positive']=data['Scores'].apply(lambda post: post['pos'])
    data['Negative']=data['Scores'].apply(lambda neg: neg['neg'])
    #creating comp_score for the final output
    data['Comp_Score']=data['Compound'].apply(lambda final_score: 'Positive' if final_score>=0 else 'Negative')
    #checking how many poitive and negative count
    posNeg=pd.DataFrame(data['Comp_Score'].value_counts())
    return posNeg,data
```

```python
posNeg,data=sentimentalAnalysis(dataset,'Chat')
```
