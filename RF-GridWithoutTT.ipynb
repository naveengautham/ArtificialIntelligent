```python
import pandas as pd
```

```python
dataset= pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset= dataset.replace({True:1,False:0})
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['squared_error','absolute_error'],'max_features':['sqrt','log2'],'n_estimators':[50,100]}
grid= GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
re= grid.cv_results_
print(re)
```

```python
table= pd.DataFrame.from_dict(re)
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_pred= grid.predict([['age_input','bmi_input','children_input','sex_male_input','smoker_yes_input']])
print('future preduiction is :', format(Future_pred))
```
